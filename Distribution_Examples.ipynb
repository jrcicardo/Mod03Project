```python
import pandas as pd
import numpy as np
import scipy as sp

from matplotlib import pyplot as plt
import seaborn as sns; sns.set()

%matplotlib inline

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

## Uniform Distribution

```python
n = 10**5
a = 100
delta = 10

data = sp.stats.uniform.rvs(size=n, 
                            loc=a, 
                            scale=delta)
```

```python
plt.figure()
ax = sns.distplot(data,
                  bins=100,
                  kde=True,
                  color='tomato',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
plt.show()
```

## Normal (Gaussian) Distribution

```python
# sample size
n = 10**5

# two parameters: mu=average, sigma=standard deviation
mu = 12
sigma = 3

data = sp.stats.norm.rvs(size=n,
                         loc=mu,
                         scale=sigma)
```

```python
data[:5]
```

```python
z_data = (data - np.mean(data))/np.std(data)
```

```python
plt.figure()
ax = sns.distplot(z_data,
                  bins=100,
                  kde=True,
                  color='tomato',
                  hist_kws={"linewidth": 15,'alpha':1})
ax = sns.distplot(data,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution ', ylabel='Frequency')
plt.show()
```

```python
data1 = sp.stats.norm.rvs(size=n, loc=0, scale=1) # standard
data2 = sp.stats.norm.rvs(size=n, loc=-5, scale=2.4) # larger std dev
data3 = sp.stats.norm.rvs(size=n, loc=3, scale=0.3) # smaller std dev

plt.figure()
ax = sns.distplot(data1,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax = sns.distplot(data2,
                  bins=100,
                  kde=True,
                  color='goldenrod',
                  hist_kws={"linewidth": 15,'alpha':1})
ax = sns.distplot(data3,
                  bins=100,
                  kde=True,
                  color='limegreen',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution ', ylabel='Frequency')
plt.show()
```

### Hacky Transforms

```python
data -= 6

plt.figure()
ax = sns.distplot(data,
                  bins=100,
                  kde=True,
                  color='tomato',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Log-Normal Distribution Transformed', ylabel='Frequency')
plt.show()
```

```python
data = np.log1p(data)

plt.figure()
ax = sns.distplot(data,
                  bins=100,
                  kde=True,
                  color='tomato',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Log-Normal Distribution Transformed', ylabel='Frequency')
plt.show()
```

```python
data = (data - np.mean(data))/np.std(data)

plt.figure()
ax = sns.distplot(data,
                  bins=100,
                  kde=True,
                  color='tomato',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Log-Normal Distribution Transformed', ylabel='Frequency')
plt.show()
```

### Transforming with SciKit

```python
from sklearn.preprocessing import QuantileTransformer as QT
```

```python
trans = QT(n_quantiles=100, output_distribution='normal')
```

```python
data = trans.fit_transform(data.reshape(-1, 1) )

plt.figure()
ax = sns.distplot(data,
                  bins=100,
                  kde=True,
                  color='tomato',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Log-Gamma Distribution Transformed', ylabel='Frequency')
plt.show()
```

### Caveat

* This is an easy way to make your data "more normal".

* It is not a "fix-all" --clearly, the closer to the tails you get, the more the data becomes abnormal.

* It can be a useful tool for implementing quick tests, but you may want to handle outliers.
